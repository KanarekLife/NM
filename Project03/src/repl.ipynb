```python
import matplotlib.pyplot as plt
from helper import *

POINTS = 100
INTERPOLATION_POINTS = 512

nodes = read_nodes("../data/chelm.csv")
```

```python
fig, axs = plt.subplots(1,2)
fig.set_size_inches(15, 5)
fig.suptitle(f'Runge phenomenon example ({POINTS} points)')

# Linspace
selected_points = find_closest_nodes(nodes, linspace(nodes[0][0], nodes[-1][0], POINTS))
x = linspace(nodes[0][0], nodes[-1][0], INTERPOLATION_POINTS)
y = lagrange_interpolation(selected_points, x)
axs[0].set_title('Lagrange interpolation (linearly spaced nodes)')
axs[0].set_yscale('log')
axs[0].plot(list(map(lambda x: x[0], nodes)), list(map(lambda x: x[1], nodes)))
axs[0].plot(x, y, color='green')
axs[0].legend(['original function', 'interpolated function'])

# Chebyshev
selected_points = find_closest_nodes(nodes, chebyshev_nodes(nodes[0][0], nodes[-1][0], POINTS))
x = linspace(nodes[0][0], nodes[-1][0], INTERPOLATION_POINTS)
y = lagrange_interpolation(selected_points, x)
axs[1].set_title('Lagrange interpolation (Chebyshev nodes)')
axs[1].set_yscale('log')
axs[0].plot(list(map(lambda x: x[0], nodes)), list(map(lambda x: x[1], nodes)))
axs[1].plot(x, y, color='green')
axs[1].legend(['original function', 'interpolated function'])

fig.savefig(f'../plots/runge_phenomenon_example.png')
```
