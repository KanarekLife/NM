```python
import matplotlib.pyplot as plt
from helper import *

POINTS = 103
INTERPOLATION_POINTS = 512

nodes = read_nodes("../data/chelm.txt")

selected_points = find_closest_nodes(nodes, linspace(nodes[0][0], nodes[-1][0], 30))
x = linspace(nodes[0][0], nodes[-1][0], INTERPOLATION_POINTS)
y = cubic_interpolation(selected_points, x)
plt.plot(list(map(lambda x: x[0], nodes)), list(map(lambda x: x[1], nodes)))
plt.plot(x, y, label="Cubic interpolation", color='green')
plt.yscale('log')
```
