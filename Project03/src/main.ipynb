```python
import matplotlib.pyplot as plt
from helper import *

POINTS = 100
INTERPOLATION_POINTS = 512

nodes = read_nodes("../data/chelm.txt")
```

```python
# Interpolation using linspace nodes
selected_points = find_closest_nodes(nodes, linspace(nodes[0][0], nodes[-1][0], POINTS))
x = linspace(nodes[0][0], nodes[-1][0], INTERPOLATION_POINTS)
y = lagrange_interpolation(selected_points, x)
plt.plot(list(map(lambda x: x[0], nodes)), list(map(lambda x: x[1], nodes)))
plt.plot(x, y, color='green')
plt.yscale('log')
```

```python
# Interpolation using Chebyshev nodes
selected_points = find_closest_nodes(nodes, chebyshev_nodes(nodes[0][0], nodes[-1][0], POINTS))
x = linspace(nodes[0][0], nodes[-1][0], INTERPOLATION_POINTS)
y = lagrange_interpolation(selected_points, x)
plt.plot(list(map(lambda x: x[0], nodes)), list(map(lambda x: x[1], nodes)))
plt.plot(x, y, color='green')
plt.yscale('log')
```
